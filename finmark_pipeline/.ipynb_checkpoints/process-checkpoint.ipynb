# Milestone 2 - Revised Data Pipeline Notebook

This notebook reads the three raw CSV files from the `data/raw` folder, validates and cleans them, keeps only the important columns for analytics use, and exports processed CSV files into the `data/processed` folder.

## Goals
- Read input files from `data/raw`
- Save cleaned files to `data/processed`
- Produce 3 processed CSV files from 3 raw CSV files
- Remove unnecessary generic columns
- Keep only the important fields needed for dashboards and analytics
- Add validation and fallback logic so the pipeline does not fail when columns are missing or messy

## 1. Import libraries

This step imports the Python libraries needed for file handling, data cleaning, schema validation, and logging. Pandas is used for transformation, while the standard library is used for path management and error handling.

```python
import os
import json
import pandas as pd
from datetime import datetime
```

## 1. Set up folders and file paths

This step defines where the raw CSV files are stored and where the processed outputs will be saved.

The notebook expects the following structure:

- `data/raw/event_logs.csv`
- `data/raw/marketing_summary.csv`
- `data/raw/trend_report.csv`

The cleaned outputs will be saved to:

- `data/processed/`

```python
BASE_DIR = Path(".")
RAW_DIR = BASE_DIR / "data" / "raw"
OUTPUT_DIR = BASE_DIR / "data" / "processed"

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

RAW_FILES = {
    "event_logs": RAW_DIR / "event_logs.csv",
    "marketing_summary": RAW_DIR / "marketing_summary.csv",
    "trend_report": RAW_DIR / "trend_report.csv",
}
```

## 2. Verify that the raw files exist

This step checks whether the notebook can find the three CSV files inside `data/raw`.

If a file is missing, the pipeline should stop early so the path problem can be fixed immediately.

```python
for name, path in RAW_FILES.items():
    print(name, "->", path, "| exists:", path.exists())
```

```python
missing_files = [str(path) for path in RAW_FILES.values() if not path.exists()]

if missing_files:
    raise FileNotFoundError(
        "The following raw files were not found:\n" + "\n".join(missing_files)
    )
else:
    print("All raw files found successfully.")
```

## 3. Define the important columns

These are the useful business columns that should remain in the processed files.

Generic columns that are not useful for analytics can be removed during processing.

```python
IMPORTANT_COLUMNS = {
    "event_logs": [
        "user_id",
        "event_type",
        "event_time",
        "product_id",
        "amount",
    ],
    "marketing_summary": [
        "date",
        "users_active",
        "total_sales",
        "new_customers",
        "report_generated",
    ],
    "trend_report": [
        "week",
        "avg_users",
        "sales_growth_rate",
    ],
}
```

## 4. Define expected data types and fallback values

This step tells the pipeline what type each important column should have.

It also defines fallback values that will be used when:
- a required column is missing, or
- a value is invalid or unreadable

```python
EXPECTED_DTYPES = {
    "event_logs": {
        "user_id": "string",
        "event_type": "string",
        "event_time": "datetime",
        "product_id": "string",
        "amount": "float",
    },
    "marketing_summary": {
        "date": "datetime",
        "users_active": "float",
        "total_sales": "float",
        "new_customers": "float",
        "report_generated": "string",
    },
    "trend_report": {
        "week": "string",
        "avg_users": "float",
        "sales_growth_rate": "float",
    },
}

FALLBACK_VALUES = {
    "string": "unknown",
    "float": 0.0,
    "datetime": "1970-01-01",
}
```

## 5. Create helper functions

These helper functions standardize column names, add missing required columns, convert data types, and keep only the columns needed for the processed outputs.

```python
def standardize_columns(df):
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace(r"[^a-z0-9_]", "", regex=True)
    )
    return df


def add_missing_columns(df, dataset_name):
    expected_cols = IMPORTANT_COLUMNS[dataset_name]
    dtype_map = EXPECTED_DTYPES[dataset_name]

    for col in expected_cols:
        if col not in df.columns:
            expected_type = dtype_map[col]
            df[col] = FALLBACK_VALUES[expected_type]

    return df


def coerce_column_types(df, dataset_name):
    dtype_map = EXPECTED_DTYPES[dataset_name]

    for col, dtype in dtype_map.items():
        if dtype == "datetime":
            df[col] = pd.to_datetime(df[col], errors="coerce")
            df[col] = df[col].fillna(pd.to_datetime(FALLBACK_VALUES["datetime"]))
        elif dtype == "float":
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(FALLBACK_VALUES["float"])
        elif dtype == "string":
            df[col] = df[col].astype("string").fillna(FALLBACK_VALUES["string"])
            df[col] = df[col].replace("<NA>", FALLBACK_VALUES["string"])

    return df


def keep_important_columns(df, dataset_name):
    return df[IMPORTANT_COLUMNS[dataset_name]].copy()
```

## 6. Create the main processing function

This function performs the core pipeline logic for one dataset:
1. Read the raw CSV file
2. Standardize column names
3. Add missing required columns
4. Convert columns into expected data types
5. Keep only the important columns
6. Remove duplicate rows
7. Save the cleaned file into `data/processed`

```python
def process_dataset(dataset_name, file_path):
    print(f"\n=== Processing: {dataset_name} ===")

    df = pd.read_csv(file_path)
    original_shape = df.shape

    df = standardize_columns(df)
    df = add_missing_columns(df, dataset_name)
    df = coerce_column_types(df, dataset_name)
    df = keep_important_columns(df, dataset_name)
    df = df.drop_duplicates()

    output_file = OUTPUT_DIR / f"{dataset_name}_cleaned.csv"
    df.to_csv(output_file, index=False)

    summary = {
        "dataset": dataset_name,
        "input_file": str(file_path),
        "output_file": str(output_file),
        "original_rows": int(original_shape[0]),
        "original_columns": int(original_shape[1]),
        "cleaned_rows": int(df.shape[0]),
        "cleaned_columns": int(df.shape[1]),
    }

    print("Saved to:", output_file)
    print("Original shape:", original_shape)
    print("Cleaned shape:", df.shape)

    return df, summary
```

## 7. Run the pipeline on all three datasets

Each file is cleaned, validated, reduced to important columns, and saved as a processed CSV file in `data/processed`.

```python
processed_data = {}
processing_summaries = []

for dataset_name, file_path in RAW_FILES.items():
    df_clean, summary = process_dataset(dataset_name, file_path)
    processed_data[dataset_name] = df_clean
    processing_summaries.append(summary)
```

This creates a JSON report that records what happened during the run.

```python
run_report = {
    "run_timestamp": datetime.now().isoformat(),
    "raw_directory": str(RAW_DIR),
    "processed_directory": str(OUTPUT_DIR),
    "datasets": processing_summaries,
}

report_path = OUTPUT_DIR / "pipeline_run_report.json"

with open(report_path, "w", encoding="utf-8") as f:
    json.dump(run_report, f, indent=4)

print("Pipeline run report saved to:", report_path)
```

```python
import os
import json
import pandas as pd
from datetime import datetime

print("Starting FinMark Resilient Data Pipeline...")

RAW_DIR = "data/raw/"
PROCESSED_DIR = "data/processed/"
os.makedirs(PROCESSED_DIR, exist_ok=True)

EXPECTED_SCHEMAS = {
    "event_logs.csv": ["event_id", "user_id", "event_type", "transaction_amount", "timestamp", "session_id", "product_id"],
    "marketing_summary.csv": ["campaign_id", "campaign_name", "clicks", "impressions", "spend", "date"],
    "trend_report.csv": ["trend_id", "keyword", "search_volume", "date"]
}

pipeline_reports = []

def validate_and_load(file_name):
    raw_path = os.path.join(RAW_DIR, file_name)
    report = {
        "timestamp": datetime.now().isoformat(),
        "file_processed": file_name,
        "status": "Success",
        "errors_found": [],
        "fallback_applied": False,
        "rows_processed": 0
    }
    
    if not os.path.exists(raw_path):
        report["status"] = "Failed"
        report["errors_found"].append(f"File missing from data/raw/: {file_name}")
        return None, report

    try:
        df = pd.read_csv(raw_path)
        report["rows_processed"] = len(df)
    except Exception as e:
        report["status"] = "Failed"
        report["errors_found"].append(f"Corrupted file format. Unable to parse CSV: {str(e)}")
        return None, report

    expected_cols = EXPECTED_SCHEMAS.get(file_name, [])
    missing_cols = [col for col in expected_cols if col not in df.columns]

    if missing_cols:
        report["status"] = "Warning"
        report["fallback_applied"] = True
        report["errors_found"].append(f"Missing critical columns: {missing_cols}")

        for col in missing_cols:
            if "amount" in col or "spend" in col or "clicks" in col or "impressions" in col or "volume" in col:
                df[col] = 0.0  # Safe numerical fallback
            else:
                df[col] = "UNKNOWN"  # Safe text classification fallback
                
    return df, report

def save_processed_data(df, base_name):
    """
    Saves datasets in 2 formats here (.csv and .parquet).
    The 3rd format (.json) is generated at the very end as the system audit log.
    """
    csv_path = os.path.join(PROCESSED_DIR, f"{base_name}_cleaned.csv")
    parquet_path = os.path.join(PROCESSED_DIR, f"{base_name}_cleaned.parquet")
    
    # Output 1: Open CSV Format Layer
    df.to_csv(csv_path, index=False)
    
    # Output 2: Optimized High-Performance Parquet Layer
    df.to_parquet(parquet_path, index=False)
    print(f"   Saved: {base_name}_cleaned.csv and {base_name}_cleaned.parquet")

print("\nProcessing Event Logs...")
df_events, report = validate_and_load("event_logs.csv")
pipeline_reports.append(report)

if df_events is not None:
    df_events = df_events.drop_duplicates()
    if 'event_type' in df_events.columns:
        df_events['event_type'] = df_events['event_type'].astype(str).str.lower()
    df_events = df_events.dropna(subset=['event_id', 'user_id'])
    save_processed_data(df_events, "event_logs")

print("\nProcessing Marketing Summary...")
df_marketing, report = validate_and_load("marketing_summary.csv")
pipeline_reports.append(report)

if df_marketing is not None:
    df_marketing = df_marketing.drop_duplicates()
    if 'campaign_name' in df_marketing.columns:
        df_marketing['campaign_name'] = df_marketing['campaign_name'].astype(str).str.strip()
    df_marketing = df_marketing.dropna(subset=['campaign_id'])
    save_processed_data(df_marketing, "marketing_summary")

print("\nProcessing Trend Report...")
df_trends, report = validate_and_load("trend_report.csv")
pipeline_reports.append(report)

if df_trends is not None:
    df_trends = df_trends.drop_duplicates()
    if 'keyword' in df_trends.columns:
        df_trends['keyword'] = df_trends['keyword'].astype(str).str.lower().str.strip()
    df_trends = df_trends.dropna(subset=['trend_id'])
    save_processed_data(df_trends, "trend_report")

report_path = os.path.join(PROCESSED_DIR, "pipeline_run_report.json")
with open(report_path, "w") as f:
    json.dump(pipeline_reports, f, indent=4)

print("\n==========================================")
print("Pipeline Run Complete!")
print("All clean data written to 'data/processed/'")
print("Audit Log Output 3 created: 'pipeline_run_report.json'")
print("==========================================")
```
